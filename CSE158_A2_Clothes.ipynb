```python
import pandas as pd
import numpy as np
from datetime import datetime

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format="retina"
```

```python
runway = pd.read_json("renttherunway_final_data.json", lines=True).dropna()
```

```python
runway.head(5)
```

```python
runway["bust size"].unique()
```

```python
# split the bust size to band and cup size
runway["band size"] = runway["bust size"].apply(lambda size: int(size[:2]))
runway["cup size"] = runway["bust size"].apply(lambda size: size[2:])
```

```python
runway.head(5)
```

```python
runway["body type"].unique()
```

```python
runway["age"].unique()
```

```python
# only keep ages that are smaller than 100
runway = runway[runway["age"] < 100]
```

```python
runway["rented for"].unique()
```

```python
# convert foot and inch to only inch
runway["height_inch"] = runway["height"].str.split().apply(lambda x: int(x[0][:-1]) * 12 + int(x[1][:-1]))
```

```python
# only keep the number without lbs
runway["weight"] = runway["weight"].str[:-3].astype(int)
```
